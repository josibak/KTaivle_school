## **2. Î∂ÑÏÑù**

- carseat Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóêÏÑú ÏïÑÎûò Î≤îÏ£ºÌòï Î≥ÄÏàòÎ•º Îã§Ïùå Î∞©Î≤ïÏúºÎ°ú Î∂ÑÏÑùÌïòÏÑ∏Ïöî.
    - ÏàòÏπòÌôî: Î≤îÏ§èÍ∞í Í∞úÏàò Î∞è ÎπÑÏú®
    - ÏãúÍ∞ÅÌôî: sns.countplot() Ìï®Ïàò, plt.pie() Ìï®Ïàò

**1Ô∏è‚É£ ShelveLoc**

```python
# Î≤îÏ£ºÍ∞í Í∞úÏàò Î∞è ÎπÑÏú® ÌôïÏù∏
temp = carseat['ShelveLoc'].value_counts()
# print(temp.index)
# print(temp.values)
print(temp)
print('='*28)
print(carseat['ShelveLoc'].value_counts(normalize=True))
```

```python
# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.bar(x=temp.index, height=temp.values)

plt.subplot(1,3,2)
sns.countplot(x='ShelveLoc', data=carseat)

plt.subplot(1,3,3)
plt.pie(x=temp.values, labels=temp.index, autopct='%.2f%%')

plt.tight_layout()
plt.show()
```

**2Ô∏è‚É£ Urban**

```python
# Î≤îÏ£ºÍ∞í Í∞úÏàò Î∞è ÎπÑÏú® ÌôïÏù∏
temp = carseat['Urban'].value_counts()
# print(temp.index)
# print(temp.values)
print(temp)
print('='*28)
print(carseat['Urban'].value_counts(normalize=True))
```

```python
# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
plt.bar(x=temp.index, height=temp.values)

plt.subplot(1,2,2)
plt.pie(x=temp.values, labels=temp.index, autopct='%.2f%%')

plt.tight_layout()
plt.show()
```

**3Ô∏è‚É£ US**

```python
# Î≤îÏ£ºÍ∞í Í∞úÏàò Î∞è ÎπÑÏú® ÌôïÏù∏
temp = carseat['US'].value_counts()
# print(temp.index)
# print(temp.values)
print(temp)
print('='*28)
print(carseat['US'].value_counts(normalize=True))
```

```python
# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
plt.bar(x=temp.index, height=temp.values)

plt.subplot(1,2,2)
plt.pie(x=temp.values, labels=temp.index, autopct='%.2f%%')

plt.tight_layout()
plt.show()
```
